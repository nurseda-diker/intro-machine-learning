```python
import pandas as pd
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
data = pd.read_csv("train.csv")
data
```

```python
X = data.drop('SalePrice',axis=1) # özellikler
y = data.loc[:,'SalePrice'] # hedefler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
linear_reg = LinearRegression()
ridge_reg = make_pipeline(StandardScaler(), Ridge(alpha=0.05)) # veriler önce standartlaştırılır ve ridge regresyon modeli oluşturulur
#Standartlaştırma, modelin performansını ve doğruluğunu artırmak için özellikle önemlidir, çünkü birçok makine öğrenimi algoritması, özelliklerin ölçeklendirilmesinden etkilenir.
#Ridge regresyon ise, modelin genelleme kabiliyetini artırarak aşırı uyumu önler.
```

```python
linear_reg.fit(X_train, y_train)
ridge_reg.fit(X_train, y_train)
```

```python
linear_pred = linear_reg.predict(X_test)
ridge_pred = ridge_reg.predict(X_test)
```

```python
linear_mse = mean_squared_error(y_test, linear_pred)
ridge_mse = mean_squared_error(y_test, ridge_pred)
```

```python
print(f"MSE without Ridge: {linear_mse}")
print(f"MSE with Ridge : {ridge_mse}")  # Ridge regresyon ile elde edilen model, MSE açısından daha iyidir
```
