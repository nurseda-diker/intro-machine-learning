```python
from sklearn.datasets import fetch_openml  # OpenML'den veri setini yüklemek için kullanılır
from sklearn.model_selection import train_test_split

X, y = fetch_openml("mnist_784", version=1, return_X_y=True, as_frame=False)
# return_X_y=True => X (özellikler) ve y (etiketler) ayrı olarak döndürülüyor.
#  as_frame=False => veri seti numpy dizisi olarak dönüyor
X = X / 255.0  # görüntü verileri 0 ile 1 arasında normalleştiriliyor

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.3)
```

```python
import pandas as pd

data = pd.DataFrame(X)  # özellikler kullanılarak bir pandas DataFrame oluşturulur.
data.insert(784, "label", y) # 784. kolona etiketler (y) eklenir ve bu kolon "label" olarak adlandırılır.
data.head()
```
