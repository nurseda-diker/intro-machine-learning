```python
import pandas as pd
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
```

```python
train.columns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
train.head()
```

```python
x_train=train.drop("SalePrice",axis=1) #axis=1 sütun , 0 satır.  SalePrice  hedef değiken olduğu için onu eğitim setinden çıkarmamız gerekiyor.
y_train=train.loc[:,"SalePrice"] #Bu satırda, train veri setinin tüm satırlarından (: ile gösterilen kısım) SalePrice sütunu seçiliyor ve y_train olarak saklanıyor.
```

```python
model.fit(x_train,y_train)
```

```python
X_test = test.drop('SalePrice', axis=1)
y_test = test.loc[:,'SalePrice']
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from numpy import sqrt
```

```python
rmse = sqrt(mean_squared_error(y_test, predictions))
rmse  # modelin tahmin hatasının ölçüsü
```

```python
comparison = pd.DataFrame({"Actual Values": y_test,"Predictions": predictions})
```
